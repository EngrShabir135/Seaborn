# Seaborn Library

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. For a brief introduction to the ideas behind the library, you can read the introductory notes or the paper.

### Import the library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load the dateset

```python
df=sns.load_dataset('tips')
```

## Relplot

The Seaborn Relational Plot (relplot) allows us to visualise how variables within a dataset relate to each other. Data visualisation is an essential part of any data analysis or machine learning workflow. It allows to gain insights about our data.

```python
sns.relplot(df.total_bill,df.tip)
```

# Hue

In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. Using the official document for lmplot provided an example for this. Adding `hue="smoker" tells seaborn you want to colour the data points for smoker and non-smoker differently.

```python
sns.relplot(df.total_bill,df.tip,hue=df.smoker)
```

# Style

Seaborn has five built-in themes to style its plots: darkgrid , whitegrid , dark , white , and ticks . Seaborn defaults to using the darkgrid theme for its plots, but you can change this styling to better suit your presentation needs. To use any of the preset themes pass the name of it to sns. set_style() .

```python
sns.relplot(df.total_bill,df.tip,hue=df.smoker,style=df.sex)
```

# The size Parameter

Size parameter is used to group by a particular column. Thus, you get multiple lines each one with different sizes.

```python
sns.relplot(df.total_bill,df.tip,hue=df.smoker,style=df.sex,size=100)
```

# Ploting with categorical data

Numerical data, as the name implies, refers to numbers. Categorical data is everything else. Categorical data is non-numerical information that is divided into groups. As its name suggests, categorical data describes categories or groups.

1) categorical scatterplots
a) stripplot
b)swarmplot

```python
import seaborn as sns

exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind",
                data=exercise)
```

a) stripplot

```python
import seaborn
import matplotlib.pyplot as plt

seaborn.set(style = 'whitegrid')
tip = seaborn.load_dataset("tips")

seaborn.stripplot(x="day", y="total_bill", data=tip)

plt.show()
```

b)swarmplot

```python
import seaborn


seaborn.set(style='whitegrid')
fmri = seaborn.load_dataset("fmri")

seaborn.swarmplot(x="timepoint",
                  y="signal",
                  data=fmri)
```

a)Pointplot

```python
# importing required packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")

# draw pointplot
sns.pointplot(x = "sex",
              y = "total_bill",
              data = data)
# show the plot
plt.show()
# This code is contributed
# by Deepanshu Rustagi.
```

b)Bar Plot

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt

# read a titanic.csv file
# from seaborn library
df = sns.load_dataset('titanic')

# who v/s fare barplot
sns.barplot(x = 'who',
            y = 'fare',
            data = df)

# Show the plot
plt.show()
```

c) CountPlot

```python
# importing the required library

import seaborn as sns
import matplotlib.pyplot as plt

# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')

# count plot on single categorical variable
sns.countplot(x ='sex', data = df)

# Show the plot
plt.show()
```
